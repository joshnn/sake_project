```python
import tokenizer

# 正規化されたデータに対する処理
# コメントの上位に来たものに含まれる単語を形態素解析して辞書を作る
def run_tokenize(comments):
  token_processed = []
  # 形態素解析エンジンにコメントを一つずつ渡して処理結果を集計する  
  for c in comments:
    token_processed.append(tokenizer.tokenize(c))
  return token_processed

# 映画単位でコメントに含まれる固有名詞のリストと頻度を取得する
tokens_by_movie_title = {}
for media_id, title, likes, comments in normalized_data['movie']:
  tokens_by_movie_title[title] = run_tokenize(comments) 

# 出来上がったデータ
# token_by_movie_title
# {
#    'titanic': {'Leonard': 100, 'Actress': 30, ... }   
# }
```
